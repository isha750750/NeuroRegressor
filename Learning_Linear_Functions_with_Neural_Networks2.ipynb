```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions and their derivatives
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Sample data
x1 = 3
x2 = 2
y_actual = 17

# Weights (w1 and w2) are initialized randomly between 1 and 10.
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)

# Hyperparameters
learning_rate = 0.01
epochs = 20

# Arrays to store error and predicted output over epochs
error_list = []
ypred_list = []

# Training loop /  The above steps are repeated for 20 epochs, adjusting the weights and refining the mode
for epoch in range(epochs):
    # Feed Forward: The prediction (ypred) is calculated by the formula ypred = x1 * w1 + x2 * w2
    z1 = x1 * w1 + x2 * w2
    y_pred = z1  # No activation, linear regression

    # The error is computed as the squared difference between the predicted and actual values
    error = (y_actual - y_pred) ** 2
    error_list.append(error)
    ypred_list.append(y_pred)

    # Back Propagation
    #  Gradients of the error with respect to the weights are computed, and the weights are updated using gradient descent.
    derror_dypred = -2 * (y_actual - y_pred)
    dypred_dw1 = x1
    dypred_dw2 = x2

    # Update weights
    grad_w1 = derror_dypred * dypred_dw1
    grad_w2 = derror_dypred * dypred_dw2

    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2

    # Print error and weights at each epoch
    print(f"Epoch {epoch + 1}/{epochs}, Error: {error:.4f}, w1: {w1:.4f}, w2: {w2:.4f}")

# Plotting the error graph / two plots are generated for visualization
plt.figure(figsize=(10, 5))

# Error plot
plt.subplot(1, 2, 1)
plt.plot(range(epochs), error_list, label='Error')
plt.title('Error Progression Across Epochs')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.grid(True)

# Predicted y plot
plt.subplot(1, 2, 2)
plt.plot(range(epochs), ypred_list, label='Predicted y', color='orange')
plt.axhline(y=y_actual, color='red', linestyle='--', label='Actual y')
plt.title('Predicted y Progression Across Epochs')
plt.xlabel('Epochs')
plt.ylabel('Predicted y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Final prediction
print(f"Final Predicted y: {ypred_list[-1]:.4f}")
```
